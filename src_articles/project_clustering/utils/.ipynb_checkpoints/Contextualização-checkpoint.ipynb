```python
import ast
import math
from textblob import TextBlob
import numpy as np
import json
from tqdm import tqdm
```

```python
def contexto(p1, p2, mapeamento):
    fx = 0 # variável que guarda o numero de documentos que p1 aparece
    fy = 0 # variável que guarda o numero de documentos que p2 aparece
    fxy = 0 # variavel qu guarda o numero de documentos que p1 e p2 aparecem em uma sentenca
    
    n = len(mapeamento[p1]) # numero de documentos
    
    for i in list(range(n)):
        if(len(mapeamento[p1][i].intersection(mapeamento[p2][i])) > 0):
            fxy += 1
        if(len(mapeamento[p1][i]) > 0):
            fx += 1
        if(len(mapeamento[p2][i]) > 0):
            fy += 1
            
    # obtendo NPMI
    npmi = 0
    factor1 = (n*fxy)/(fx*fy)
    factor2 = (fxy/n)
    
    try:
        value = math.log10(factor1)/-math.log10(factor2)
        npmi = (value+1)/2 # normalizando para o intervao [0,1]
    except:
        npmi = 0.0
        
    return npmi
```

```python
#Diretorios dos arquivos de sentencas
files_sentences = ['sentencas-camera.txt','sentencas-cells.txt','sentencas-dvds.txt','sentencas-laptops.txt','sentencas-routers.txt']
#Diretorios dos arquivos de palavras
files_words = ['palavras-camera.json','palavras-cells.json','palavras-dvds.json','palavras-laptops.json','palavras-routers.json']
#Diretorios dos arquivos de contexto
save_context = ['contexto-camera.json','contexto-cells.json','contexto-dvds.json','contexto-laptops.json','contexto-routers.json']
#Diretorios dos arquivos de contextualizacao
files_context = ['contextualizacao-camera.json','contextualizacao-cells.json','contextualizacao-dvds.json','contextualizacao-laptops.json','contextualizacao-routers.json']
```

```python
for index, _ in enumerate(files_sentences):
    documents = [] # Documetnos do dominio especifico
    words = {} # variavel que mapeia palavras para sentencas
    
    with open('../../datasets_processed/' + files_words[index], 'r') as fp:
        data_words = json.load(fp)
        
    sentences = [] # varivael que guarda as sentencas
    arq = open('../../datasets_processed/sentencas/' + files_sentences[index], 'r')
    for line in arq.readlines():
        if(line != '\n'):
            sentences.append(line.lower())
        else:
            documents.append(sentences)
            sentences = []
    arq.close()
    
    keys_words = sorted(data_words.keys())
    
    with open('../../datasets_processed/contextualizacao/' + files_context[index], 'r') as fp:
        data_context = json.load(fp)
        
    for key in data_context.keys():
        data_context[key] = [set(v) for v in data_context[key]]
        
    tam = len(keys_words) # quantidade de palavras no gabarito
    total = (1+tam)*tam/2 # totalde iterações que serão feitas
    cont = 0 #contatdor
    percent = 0 # valor da porcebtagem do processamento
    aux = 0 # variavel auxiliar
    
    # slvando o valor do NPMI entre as palavras
    with open('../../datasets_processed/contextos/' + save_context[index], 'w') as fp:
        data_saved = {}
        for i in tqdm(range(len(keys_words)), desc='1st loop'):
            ctx_value = {}
            for j in tqdm(range(i+1, len(keys_words)), desc='2st loop'):
                cont += 1
                value = contexto(keys_words[i], keys_words[j], data_context)
                ctx_value[keys_words[j]] = value
                
                aux = int(cont*100/total)
                if(aux > percent):
                    percent = aux
            data_saved[keys_words[i]] = ctx_value
        
        json.dump(data_saved, fp)
```
